```
# Importing the Keras libraries and packages
import numpy as np
from keras.models import Sequential
from keras.layers import Conv2D
from keras.layers import MaxPooling2D
from keras.layers import Flatten
from keras.layers import Dense
```

```
# Read Images from Folders      
from keras.preprocessing.image import ImageDataGenerator

train_datagen = ImageDataGenerator(rescale = 1./255)
training_set = train_datagen.flow_from_directory('chest_xray/train',
                                                 target_size = (64, 64),
                                                 batch_size = 32,
                                                 class_mode = 'binary')

test_datagen = ImageDataGenerator(rescale = 1./255)
test_set = test_datagen.flow_from_directory('chest_xray/test',
                                            target_size = (64, 64),
                                            batch_size = 32,
                                            class_mode = 'binary')   
from matplotlib import pyplot as plt
x,y = training_set.next()
for i in range(0,10):
    image = x[i]
    plt.imshow(image)
    plt.show()
```

```
# Initialising the CNN
classifier = Sequential()

# Step 1 - Convolution
classifier.add(Conv2D(32, (3, 3), input_shape = (64, 64, 3), activation = 'relu'))

# Step 2 - Pooling
classifier.add(MaxPooling2D(pool_size = (2, 2)))

# Step 3 - Convolution
classifier.add(Conv2D(32, (3, 3), activation = 'relu'))

# Step 4 - Pooling
classifier.add(MaxPooling2D(pool_size = (2, 2)))

# Step 5 - Convolution
classifier.add(Conv2D(32, (3, 3), activation = 'relu'))

# Step 6 - Flattening
classifier.add(Flatten())

# Step 7 - Full connection layer
classifier.add(Dense(units = 128, activation = 'relu'))

# Step 8 - Output Layer
classifier.add(Dense(units = 1, activation = 'sigmoid'))

# Compiling the CNN
classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])

# Show Model Summary
classifier.summary()
```

```
#Fitting the CNN to the images
history = classifier.fit_generator(training_set,
                         steps_per_epoch = 5216,
                         epochs = 5,
                         validation_data = test_set,
                         validation_steps = 624)

# save model to single file
classifier.save("ChestxrayModel.hdf5") # Hierarchical Data Format Fifth Version
print("Saved model to disk")

# save training history to another file
import pickle 
filehandler = open('history.pkl', 'wb') 
pickle.dump(history, filehandler)
```

```
from keras.models import load_model
classifier = load_model('ChestxrayModel.hdf5')

# Evaluate Model
print(classifier.evaluate(training_set))
"""
[8.231095804944744e-10, 1.0] # Training Accuracy: 100%
"""
print(classifier.evaluate(test_set))
"""
[8.829236030578613, 0.7724359035491943] # Test Accuracy: 77.2%
"""
test_set.reset()
y_pred = classifier.predict_generator(test_set)
y_test = test_set.classes[test_set.index_array]
y_pred = ((y_pred>0.5)*1.0).flatten()  # Set the threshold to be 0.5
print(sum(y_pred==y_test)/y_pred.size) # Test Accuracy: 0.772

# Confusion Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)

"""
[[93 141
  1  389]]
"""
```
